```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding='ISO-8859-1')
star_wars = star_wars[star_wars["RespondentID"].notnull()]
star_wars.head(10)
```

```python
yes_no = {"Yes":True, "No":False}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

star_wars.head(10)
```

```python
old_col_names = list(star_wars.columns[3:9])
#Makes a list of the old column names
new_col_names = []
new_col_names_map = {}

movie_titles = []
seen_movie_map = {}


for i in range(0,6):
    new_col_names.append("seen_{0}".format(i+1))
    #Makes a list of the new column names
    new_col_names_map[old_col_names[i]] = new_col_names[i]
    #Creates a map of the old column to new column names
    
    movie_titles.append(star_wars.iloc[0][i+3])
    #Makes a list of all the movies
    seen_movie_map[movie_titles[i]] = True

star_wars = star_wars.rename(columns=new_col_names_map)

for col_name in new_col_names:
    star_wars[col_name] = star_wars[col_name].map(seen_movie_map)
    
pd.set_option('display.max_columns', None)
#To display all the columns
star_wars.head(10)
```

```python
#Converting the columns to a float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

col_names_9_15 = list(star_wars.columns[9:15])
new_col_names_9_15 = []
col_names_9_15_map = {}

#Renaming the column headers for columns 9-15
for i in range(len(col_names_9_15)):
    new_col_names_9_15.append("ranking_{0}".format(i+1)) 
    col_names_9_15_map[col_names_9_15[i]] = new_col_names_9_15[i]
    
star_wars = star_wars.rename(columns=col_names_9_15_map)

star_wars
    
```

```python
#Finding the mean of each column

means = star_wars[new_col_names_9_15].mean()

import matplotlib.pyplot as plt
%matplotlib inline

#plotting the means of each movie
means.plot(kind="Bar")
```

# Movie Rankings

Note that these are rankings, so the lower the rating the better.

Rankings 1,2,3 are the prequels, and are all rated lower than movies 4,5,6, which are the originals. The overall favourite is movie number 5 (The Empire Strikes Back).

```python
#Finding out how many people have seen each movie.

seen_movie_count = star_wars[star_wars.columns[3:9]].sum()
seen_movie_count.plot(kind="Bar")
```

# How many people have seen each movie

Despite the originals being ranked better than the originals, the number of people who saw the first prequel is higher than the original movie. This is likely due to the increased advertising and general anticipation for the return of the series after a long hiatus. This drops off for the second and third sequel, likely due to people being disapointed by the first prequel.

```python
#Finding the difference between Star Wars and non Star Wars fans
#How many people have seen each movie
star_wars_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]==True]
not_star_wars_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]==False]

fans_movie_count = star_wars_fans[star_wars_fans.columns[3:9]].sum()
not_fans_movie_count = not_star_wars_fans[not_star_wars_fans.columns[3:9]].sum()

fans_movie_count.plot(kind="Bar")
not_fans_movie_count.plot(kind="Bar",color="Green")
```

```python
#The means of movie rankings between fans and non-fans

means_fans = star_wars_fans[new_col_names_9_15].mean()
means_not_fans = not_star_wars_fans[new_col_names_9_15].mean()

figure = plt.figure()
ax1 = figure.add_subplot(1,2,1)
ax1.bar(range(len(means_fans)), means_fans)
ax2 = figure.add_subplot(1,2,2)
ax2.bar(range(len(means_not_fans)), means_not_fans, color="Green")

plt.show()
```

# Results

There are significant differences in total numbers of who has seen the series between fans and non-fans. Fans tend to rate the original 3 movies the highest, while non-fans have a more even distribution in movie ratings.
